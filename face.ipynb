```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_lfw_people
import seaborn as sns
import cv2
```

```python
images = fetch_lfw_people(min_faces_per_person = 90)
plt.figure(figsize=(8,8))
labels = images.target
for i in range(16):
    plt.subplot(4,4,i+1)
    plt.imshow(images.images[i],cmap='gray')
    plt.title(images.target_names[labels[i]])
    plt.axis('off')
```
